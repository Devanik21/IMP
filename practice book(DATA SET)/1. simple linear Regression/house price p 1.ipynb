```python
from pandas import *
from seaborn import *
from matplotlib.pyplot import *
from sklearn.linear_model import LinearRegression
```

```python
df=read_csv('homeprices.csv')
```

```python
df
```

```python
x= df.iloc[:, 0:1].values       # iloc[ : ,:-1]
y = df.iloc[:, 1].values
```

```python
plot(x,y)
xlabel('area')
ylabel('price')
```

```python
scatter(x,y)
xlabel('area')
ylabel('price')
```

```python
r=LinearRegression()
r.fit(x,y)
```

```python
scatter(x,y,color='blue')
plot(x,r.predict(x))
show()
```

let's extend the accuracy by polynomial regression 

```python
from sklearn.preprocessing import PolynomialFeatures
r2=PolynomialFeatures(degree=2)  #overall  - lower degree is versatile 
x2=r2.fit_transform(x)           #For area <=4500, degree 5 is best
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x2, y)
```

```python
lin_reg_2.score(x2,y)
```

```python
pred= lin_reg_2.predict(r2.fit_transform([[5000]]))
pred
```
