```python
from numpy import *
from matplotlib.pyplot import *
from pandas import *
from seaborn import*
%matplotlib inline
```

```python
df=read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df
```

```python

x = df.iloc[:, 1:2].values
y = df.iloc[:, 2].values
```

```python
violinplot(x)
```

```python
violinplot(y,color='violet')
```

```python
from sklearn.linear_model import LinearRegression
r=LinearRegression()
r.fit(x,y)
```

```python
r.score(x,y)
```

```python
# Step 3 - Visualize Linear Regression Results

scatter(x, y,color = 'red')
plot(x, r.predict(x))
title("Linear Regression")

```

```python
pred =r.predict([[74]])
```

```python
pred
```

```python
from sklearn.preprocessing import PolynomialFeatures
r2=PolynomialFeatures(degree=10)
X2=r2.fit_transform(x)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X2, y)
```

```python
lin_reg_2.score(X2,y)
```

```python

scatter(x, y, color="Black")
plot(x, lin_reg_2.predict(X2))
title("Poly Regression of Degree 10")
```

```python
pred20= lin_reg_2.predict(r2.fit_transform([[80]]))
pred20
```
